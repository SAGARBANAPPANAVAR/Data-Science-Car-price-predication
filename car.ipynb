```python
finaldata['no of years'] = finaldata['Current_year'] - finaldata['Year']
```

```python
finaldata.drop(['Year'],axis = 1, inplace = True)
```

```python
finaldata.drop(['Current_year'],axis = 1, inplace = True)
```
